# Partie A — Importation du Dataset, Arrays, Premières Manipulations

### A1. Importation du dataset
Importez NumPy et Matplotlib.  
Chargez le fichier `iris.csv` avec `np.genfromtxt` (`delimiter=','`, `skip_header=1`).  
Stockez les données dans un array nommé `data` et affichez sa `shape`.


```python
# Votre code ici
import numpy as np
import matplotlib.pyplot as plt

# A1. Importation du dataset
data = np.genfromtxt('Iris.csv', delimiter=',', skip_header=1)
print("Shape de data :", data.shape)
```

### A2. Séparation des features et des espèces
Séparez :
- les 4 premières colonnes numériques dans un array `features`
- la dernière colonne (espèces) dans un array `species`
Utilisez le slicing.


```python
# Votre code ici
# Séparation des features et des espèces
features = data[:, :4]   # colonnes 0 à 3
species = data[:, 4]     # dernière colonne
print("features shape :", features.shape)
print("species shape :", species.shape)
```

### A3. Indexing
Affichez les 5 premières lignes de `features` en utilisant l’indexing.


```python
# Votre code ici
# Indexing : 5 premières lignes de features
print(features[:5])
```

Créez un array `sepal_length` contenant uniquement la première colonne de `features`.

```python
# Votre code ici
# Extraction de la première colonne
sepal_length = features[:, 0]
print("size :", sepal_length.size)
print("dtype :", sepal_length.dtype)
```

### A5. Slicing par espèce

# Partie B — Statistiques Basiques et Opérations Mathématiques

### B6. Moyenne par colonne
Calculez la moyenne de chaque colonne de `features` avec `np.mean(axis=0)`.


```python
# Votre code ici
# Moyenne par colonne
mean_features = np.mean(features, axis=0)
print("Moyenne par colonne :", mean_features)
```

### E23. Grille 2x2
Histogrammes des 4 caractéristiques avec `plt.subplots(2,2)`.


```python
# Votre code ici
# Grille 2×2 d’histogrammes des 4 caractéristiques
fig, axes = plt.subplots(2, 2, figsize=(9, 7))
labels = ['Sepal length', 'Sepal width', 'Petal length', 'Petal width']

for i, ax in enumerate(axes.flat):
    ax.hist(features[:, i], bins=15, edgecolor='black', alpha=0.7)
    ax.set_title(labels[i])
    ax.set_xlabel('cm')
    ax.set_ylabel('Fréquence')
    ax.grid(axis='y')

plt.tight_layout()
plt.show()
```

### E24. Bar chart avec error bars
Moyennes par espèce avec écart-type.


```python
# Votre code ici
# Bar chart moyennes + écart-type par espèce
species_labels = ['Setosa', 'Versicolor', 'Virginica']
colors = ['skyblue', 'lightgreen', 'plum']

fig, ax = plt.subplots(figsize=(8, 4))
x = np.arange(4)  # 4 caractéristiques
width = 0.25

for i, (start, end, lab, col) in enumerate(zip([0, 50, 100],
                                               [50, 100, 150],
                                               species_labels,
                                               colors)):
    block = features[start:end]
    means = block.mean(axis=0)
    stds  = block.std(axis=0)
    ax.bar(x + i*width, means, width, yerr=stds,
           label=lab, color=col, capsize=3)

ax.set_xlabel('Caractéristiques')
ax.set_ylabel('Moyenne (cm)')
ax.set_xticks(x + width)
ax.set_xticklabels(['Sep L', 'Sep W', 'Pet L', 'Pet W'])
ax.legend()
ax.grid(axis='y')
plt.show()
```

```python
# Votre code ici
# Line plot superposé : cumsum des largeurs des pétales par espèce
plt.figure(figsize=(7, 4))
for start, end, lab, col in zip([0, 50, 100],
                                [50, 100, 150],
                                species_labels,
                                ['tab:blue', 'tab:green', 'tab:red']):
    pet_w = features[start:end, 3]
    plt.plot(np.cumsum(pet_w), label=lab, color=col, marker='o', markersize=3)
plt.xlabel('Indice de la fleur (par espèce)')
plt.ylabel('Cumsum petal width')
plt.title('Cumsum des largeurs de pétales par espèce')
plt.legend()
plt.grid(True)
plt.show()
```

# Partie F — Mini-Analyse Finale

### F27. Corrélations
Calculez la matrice de corrélation avec `np.corrcoef`.


```python
# Votre code ici
# Matrice de corrélation (lignes = variables)
corr = np.corrcoef(features, rowvar=False)  # 4×4
print("Matrice de corrélation :\n", corr.round(2))
```

### F28. Visualisation des corrélations
Visualisez les corrélations sous forme de bar chart ou scatter.


```python
# Votre code ici
# Bar chart des corrélations Setosa (exemple)
set_corr = np.corrcoef(setosa_features, rowvar=False).flatten()
labels = [f"{i}-{j}" for i in range(4) for j in range(4)]
plt.figure(figsize=(8, 3))
plt.bar(labels, set_corr)
plt.xticks(rotation=45)
plt.ylabel("Corrélation")
plt.title("Corrélations – Setosa")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### F29. Variance par espèce
Identifiez l’espèce avec la plus grande variance globale et visualisez ses distributions.


```python
# Votre code ici
# Variance globale par espèce
vars_global = []
for start, end, lab in zip([0, 50, 100], [50, 100, 150], species_labels):
    block = features[start:end]
    vars_global.append(block.var(axis=0).sum())   # somme des variances des 4 variables
vars_global = np.array(vars_global)

max_var_idx = vars_global.argmax()
print("Espèce la plus variable :", species_labels[max_var_idx])

# Histogramme des 4 caractéristiques pour cette espèce
most_var_block = features[max_var_idx*50 : (max_var_idx+1)*50]
plt.figure(figsize=(8, 4))
for i, lab in enumerate(['Sep L', 'Sep W', 'Pet L', 'Pet W']):
    plt.subplot(1, 4, i+1)
    plt.hist(most_var_block[:, i], bins=15, color='skyblue', edgecolor='k')
    plt.title(lab)
plt.suptitle(f'Distributions – {species_labels[max_var_idx]}')
plt.tight_layout()
plt.show()
```
